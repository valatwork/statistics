## Set operations

#### Union ($A \cup B$)
A **union** of events $A,B,C,...$  is an event consisting of *all* the outcomes in all these events.

It occurs if *any* of the events $A,B,C,...$ occurs.

*OR*: $A$ or $B$ or $C$ or ...

#### Intersection ($A \cap B$)
An **intersection** of events $A,B,C,...$ is an event consisting of *only* the outcomes that are *common* in all these events.

It occurs if *each* $A,B,C,...$ occurs.

*AND*: $A$ and $B$ and $C$ and ...


#### Complement ($\bar{A}$ or $A^c$)
A **complement** of an event $A$ is an event that occurs every time when $A$ does not occur.

It consists of all outcomes excluded from $A$.

*NOT*: not $A$

#### Difference ($A \setminus B$)

A **difference** of events $A$ and $B$ consists of all outcomes included in $A$ but excluded from $B$.

*BUT NOT*: $A$ but not $B$

```python
from matplotlib_venn import venn2, venn2_circles
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Set up the figure and subplots
plt.figure(figsize=(8, 8))

# Union
plt.subplot(2, 2, 1)
plt.title('Union')
venn = venn2(subsets=(3, 3, 2), set_labels=('A', 'B'),  set_colors=('skyblue', 'skyblue'), alpha=0.5)
venn2_circles(subsets=(3, 3, 2), linestyle='solid')
venn.get_label_by_id('10').set_text('')
venn.get_label_by_id('01').set_text('')
venn.get_label_by_id('11').set_text('A ∪ B')
venn.get_patch_by_id('11').set_color('skyblue')


# Intersection
plt.subplot(2, 2, 2)
plt.title('Intersection')
venn = venn2(subsets=(3, 3, 2), set_labels=('A', 'B'), set_colors=('white', 'white'))
venn2_circles(subsets=(3, 3, 2), linestyle='solid')
venn.get_label_by_id('10').set_text('')
venn.get_label_by_id('01').set_text('')
venn.get_label_by_id('11').set_text('A ∩ B')
venn.get_patch_by_id('11').set_color('skyblue')


# Complement
plt.subplot(2, 2, 3)
plt.title('Complement')
plt.gca().add_patch(plt.Rectangle((-1, -1), 2, 2, fill=True, color='skyblue', alpha=0.5, zorder=0))
venn = venn2(subsets=(1, 0, 0), set_labels=('', r'$\bar{A}$'), set_colors=('white', 'white'))
venn2_circles(subsets=(1, 0, 0), linestyle='solid')
venn.get_label_by_id('10').set_text('A')
venn.get_label_by_id('01').set_text('')
# Draw a white circle on top with the same radius and position
circle = plt.Circle((0, 0), 0.55, color='white', zorder=0.1)
plt.gca().add_patch(circle)

# Difference
plt.subplot(2, 2, 4)
plt.title('Difference')
venn = venn2(subsets=(3, 3, 2), set_labels=('A', 'B'), set_colors=('skyblue', 'white'))
venn2_circles(subsets=(3, 3, 2), linestyle='solid')
venn.get_label_by_id('10').set_text(r'$A \setminus B$')
venn.get_label_by_id('01').set_text('')
venn.get_label_by_id('11').set_text('')


# Show the plots
plt.tight_layout()
plt.show()
```

Events $A$ and $B$ are **disjoint** if their intersection is empty:
$$A \cap B = \emptyset$$

Events $A_1, A_2, A_3, ...$ are **mutually exclusive** or **pairwise disjoint** if any two of these events are disjoint:

$$A_i \cap A_j = \emptyset \quad for \ any \quad  i \neq j$$

Events $A, B, C, ...$ are **exhaustive** if their union is the whole sample space:
$$A \cup B \cup C \cup ... = \Omega$$

Mutually exclusive events will never occur at the same time. Occurrence of any one of them eliminates the possibility for all the others to occur.

Exhaustive events cover the entire $\Omega\$, so that “there is nothing left.” In other words, among any collection of exhaustive events, at least one occurs for sure.
