# Independent vs Dependent Variables

## Conventions

- Dependent variable: the variable that you are trying to predict or explain
    - *Synonyms: label, response variable, output, outcome, or target*
- Typically denoted with $y$
- Cartesian vertical axis

- Independent variable: the variable that is used to predict or explain the dependent variable
    - *Synonyms: explanatory variable, predictor, input, or feature.*
- May *predict* or *explain* the outcome
- Typically denoted with $x$
- Cartesian horizontal axis
- Ideally can be explicitly adjusted, not only measured

### Visual representation with Pyhon

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = np.array([0, 1, 2, 3, 4, 5, 6, 7.])
y = np.array([1.86, 1.31, .62, .33, .09, -.67, -1.23, -1.37])
sns.scatterplot(x=x, y=y)
plt.title("Independent vs Dependent Variables")
plt.xlabel(r"$x$ (Independent Variable)")
_ = plt.ylabel(r"$y$ (Dependent Variable)")
```
