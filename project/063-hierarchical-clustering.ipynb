# Hierarchical Clustering

The two main approaches to hierarchical clustering are *agglomerative* and *divisive* hierarchical clustering.

In divisive hierarchical clustering, we start with one cluster that encompasses the complete dataset, and we iteratively split the cluster into smaller clusters until each cluster only contains one example.

Agglomerative clustering takes the opposite approach.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./data/cluster-mpg.csv')
```

```python
df
```

```python
df.describe()
```

```python
df['origin'].value_counts()
```

```python
df_w_dummies = pd.get_dummies(df.drop('name', axis=1))
```

```python
df_w_dummies
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaledr = MinMaxScaler()
```

```python
scaled_data = scaledr.fit_transform(df_w_dummies)
```

```python
scaled_data
```

```python
scaled_df = pd.DataFrame(scaled_data, columns=df_w_dummies.columns)
```

```python
scaled_df
```

```python
sns.heatmap(scaled_df)
```

```python
sns.clustermap(scaled_df)
```

```python
sns.clustermap(scaled_df,row_cluster=False)
```

```python
sns.clustermap(scaled_df,col_cluster=False)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=4)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
sns.scatterplot(data=df,x='mpg',y='weight',hue=cluster_labels)
```

```python
model = AgglomerativeClustering(n_clusters=None,distance_threshold=0)
```

```python
cluster_labels = model.fit_predict(scaled_df)
```

```python
cluster_labels
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy
```

```python
linkage_matrix = hierarchy.linkage(model.children_)
```

```python
linkage_matrix
```

```python
plt.figure(figsize=(20,10))
dn = hierarchy.dendrogram(linkage_matrix)
```

```python
plt.figure(figsize=(20,10))
dn = hierarchy.dendrogram(linkage_matrix,truncate_mode='lastp',p=48)
```
