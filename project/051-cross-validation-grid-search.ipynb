### Cross Validation with cross_val_score

```python
import pandas as pd
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('../project/data/another_advertising.csv')
```

```python
df.head()
```

```python
X = df.drop('sales', axis=1)
y = df['sales']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = Ridge(alpha=100)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(model, X_train, y_train, scoring='neg_mean_squared_error', cv=5)
```

```python
scores
```

```python
abs(scores.mean())
```

```python
model = Ridge(alpha=1)
```

```python
scores = cross_val_score(model, X_train, y_train, scoring='neg_mean_squared_error', cv=5)
```

```python
abs(scores.mean())
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mean_squared_error(y_test, y_pred)
```

### Cross Validation with cross_validate

The `cross_validate` function differs from `cross_val_score` in two ways:

1. It allows specifying multiple metrics for evaluation.
2. It returns a dictionary containing fit-times, score-times in addition to test score.

```python
X = df.drop('sales', axis=1)
y = df['sales']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = Ridge(alpha=100)
```

```python
from sklearn.model_selection import cross_validate
```

```python
scores = cross_validate(model,X_train,y_train,
                         scoring=['neg_mean_absolute_error','neg_mean_squared_error','max_error'],cv=5)
```

```python
scores
```

```python
pd.DataFrame(scores)
```

```python
pd.DataFrame(scores).mean()
```

```python
model = Ridge(alpha=1)
```

```python
scores = cross_validate(model,X_train,y_train,
                         scoring=['neg_mean_absolute_error','neg_mean_squared_error','max_error'],cv=5)
```

```python
pd.DataFrame(scores).mean()
```

## Grid Search

A grid search is a way of training and validating a model on every possible combination of multiple hyperparameters (provided as a dictionary) to find the best combination.

Sci-kit learn provides a way to do this with the `GridSearchCV` class.

```python
X = df.drop('sales',axis=1)
y = df['sales']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNet
```

```python
base_elastic_model = ElasticNet()
```

```python
param_grid = {'alpha':[0.1,1,5,10,50,100],
              'l1_ratio':[.1, .5, .7, .9, .95, .99, 1]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(estimator=base_elastic_model,
                          param_grid=param_grid,
                          scoring='neg_mean_squared_error',
                          cv=5,
                          verbose=1)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_params_
```

```python
pd.DataFrame(grid_model.cv_results_)
```

```python
y_pred = grid_model.predict(X_test)
```

```python
mean_squared_error(y_test,y_pred)
```
