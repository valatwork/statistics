# Exponential and Laplace distributions

## Exponential distribution

An exponential distribution models a continuous random variable over time, area or space where the rate of occurrences decreases as X gets larger.

The exponential distribution is often concerned with the amount of time until some specific event occurs.

The random variable for the exponential distribution is continuous and often measures a passage of time, although it can be used in other applications. Typical questions may be, “what is the probability that some event will occur within the next  $x$ hours or days, or what is the probability that some event will occur between $x_1$ hours and $x_2$ hours, or what is the probability that the event will take more than $x_1$ hours to perform?”

In more practical terms: product lifetime, the time until the next earthquake, the time until a radioactive particle decays, the time between phone calls, etc.

It is often used in deep learning models.

### Memoryless Property of the Exponential Distribution


The memorylessness property of the exponential distribution says that, for an exponentially distributed random variable $X$ and any nonnegative times $s$ and $t$, the probability that $X$ exceeds $s + t$ given that it already exceeds $s$ is the same as the probability that $X$ exceeds $t$. Symbolically,
$$\Pr(X > s + t \,\mid\, X > s) \;=\; \Pr(X > t)$$

Once you reach time $s$ without $X$, you “forget” about the past, and the probability of waiting at least another $t$ units is the same as if you started waiting from scratch.

In short: the knowledge of what has occurred in the past has no effect on future probabilities.


#### Plotting the exponential distribution

Relatively squished up against zero and then decreases exponentially toward higher values. Log-normal distribution, in contrast, could take negative values and both increases and decreases.

```python
import numpy as np
import seaborn as sns
```

```python
rng = np.random.default_rng()
x = rng.exponential(scale=4, size=10000)
```

```python
_ = sns.displot(x)
```

Its logarithm has a skewed distribution:



```python
_ = sns.displot(np.log(x), color='indigo', kde=True)
```

## Laplace distribution

The Laplace distribution is the distribution of the difference of two independent, identically distributed exponential variables.

For this reason, it is also called the *double exponential distribution*.

Real-world applications:
- Extreme weather events
- Machine learning
- Image processing
- Signal processing

```python
x = rng.laplace(size=10000)
```

```python
_ = sns.displot(x, kde=True)
```

The `loc` parameter shifts the distribution along the x-axis and allows it to be centered on any value

```python
_ = sns.displot(rng.laplace(loc=25, size=10000), kde=True)
```
