# Bernoulli, Binomial and Multinomial distributions

## Bernoulli distribution

A random variable with two possible values, 0 and 1, is called a Bernoulli variable, its distribution is Bernoulli distribution, and any experiment with a binary outcome is called a Bernoulli trial.

The Bernoulli distribution is the simplest discrete distribution taking just two values $0$ and $1$, with probabilities $1 − p$ and $p$, respectively.

##### pdf of $X$

$$P{\mathrm{(X = 0)}} = 1 - p$$

$$P{\mathrm{(X = 1)}} = p$$

Where $0 \leq p \leq 1$.

#### cdf

$$F{\mathrm{(x)}} = \begin{cases} 0 & \text{if } x < 0 \\ 1 - p & \text{if } 0 \leq x < 1 \\ 1 & \text{if } x \geq 1 \end{cases}$$

Good or defective components, parts that pass or fail tests, transmitted or lost signals,
working or malfunctioning hardware, benign or malicious attachments, sites that contain
or do not contain a keyword, heads and tails, and so on, are examples of
Bernoulli trials.

## Binomial distribution

The probability distribution for a *discrete* variable that meets these criteria:

- The variable is for a sample that consists of a fixed number of experimental trials (the sample size).
- The variable has only two mutually exclusive and collectively exhaustive events, typically labeled as success and failure.
- The probability of an event being classified as a success, $p$, and the probability of an event being classified as a failure, $1 – p$, are both constant in all experimental trials.

For example, consider an experiment made up of a repeated number of independent and identical trials having only two outcomes, like tossing a fair coin. These experiments with only two possible outcomes on each trial are called *Bernoulli trials*. Often, they are found by assigning either an $S$ (success) or $F$ (failure) or a $0$ or $1$ to an outcome. Something either happened (1) or did not happen (0).

### Example with Python

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng()
n = 5
n_experiments = 1000
heads_count = rng.binomial(n, 0.5, n_experiments)
heads, event_count = np.unique(heads_count, return_counts=True)
event_proba = event_count/n_experiments
plt.bar(heads, event_proba)
plt.xlabel('Heads flips (out of 5 tosses)')
_ = plt.ylabel('Event probability')
plt.title('Binomial distribution with p=0.5')
```

```python
heads_count = np.random.binomial(n, 0.8, n_experiments)
heads, event_count = np.unique(heads_count, return_counts=True)
event_proba = event_count/n_experiments
```

```python
plt.bar(heads, event_proba)
plt.xlabel('Heads flips (out of 5 tosses)')
_ = plt.ylabel('Event probability')
plt.title('Binomial distribution with p=0.8')
```

## Multinomial distribution

Generalization of the binomial distribution to discrete random variables with more than two possible outcomes, e.g., the roll of a die.

Since multinomial distributions can represent any categorical variable (e.g., burger, hot dog, pizza; 52 playing cards; NFL teams), not just integer values, we can't always calculate an expected value.

```python
n = 10000
rolls = rng.multinomial(n, [1/6]*6)
rolls
```

```python
event_proba = rolls/n
```

```python
plt.bar(range(1,7), event_proba)
plt.xlabel('Die roll')
_ = plt.ylabel('Event probability')
plt.title('Multinomial distribution with 10000 rolls')
```


As in the binomial case, multinomial distributions are created by sampling $n > 1$ multinoulli distributions, where the multinoulli distribution is equivalent to the multinomial distribution when $n = 1$.
