## Covariance

A measure of the extent to which one variable varies in concert with another (i.e., similar magnitude and direction).



### Formula

One way to quantify the relationship between two variables is the covariance. Remember that variance for a single variable is computed as the average squared difference between each data point and the mean:

$$
\sigma^2 = \frac{\sum_{i=1}^{N} (x_i - \mu)^2}{N}
$$

This tells us how far each observation is from the mean, on average, in squared units. Covariance tells us whether there is a relation between the deviations of two different variables across observations. It is defined as:

$$ cov(x, y) = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{n-1} $$

Note: with large datasets, the degree of freedom can be set to n instead of n-1.

### Covariance in Python

#### Positive covariance

```python
import numpy as np
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
x = iris.sepal_length
y = iris.petal_length
_ = sns.scatterplot(x=x, y=y)
```

By default, `np.cov` uses the formula for the population covariance, which divides by `n` instead of `n-1`. This is the same as setting `ddof=0`.

The method will return a *covariance matrix*:

```python
np.cov(x, y, ddof=0)
```

To get the covariance between the two variables, we can access the first row and first column of the covariance matrix:

```python
np.cov(x, y, ddof=0)[0,1]
```

The diagonal elements of the covariance matrix are the variances of the variables, while the off-diagonal elements are the covariances between the variables.

```python
np.var(x) , np.var(y)
```

#### Negative covariance

If $x$ and $y$ are *inversely related*, their covariance is negative. The less related they are, the closer their covariance is to zero:

```python
_ = sns.scatterplot(x=iris.sepal_length, y=iris.sepal_width)
```

```python
np.cov(iris.sepal_length, iris.sepal_width, ddof=0)
```
